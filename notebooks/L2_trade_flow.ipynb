# Layer 2: Trade Flow Anatomy


```python
import os
import numpy as np
import pandas as pd

from mda.loader import load_trades
from mda.timestamps import add_ts_columns
from mda import EXCHANGES, DATA_DIR
from mda.layer2.sizes import compute_size_distribution, compute_notional_distribution
from mda.layer2.volleys import detect_volleys, volley_size_distribution
from mda.layer2.aggressor import (
    compute_aggressor_imbalance,
    compute_overall_buy_ratio,
    compute_tick_frequency,
)
from mda.plots.charts import (
    save_figure,
    plot_trade_size_histogram,
    plot_volley_distribution,
    plot_aggressor_imbalance,
    plot_tick_frequency,
)
```

```python
DATA_DIR = "/data/parquet"
REPORTS_DIR = "/data/reports"
os.makedirs(REPORTS_DIR, exist_ok=True)

df = load_trades(DATA_DIR, exchanges=EXCHANGES, add_ts_cols=True)
print("Shape:", df.shape)
print(df.dtypes)
```

```python
size_dist = compute_size_distribution(df)
notional = compute_notional_distribution(df)

print("Size distribution per exchange:")
print(size_dist["per_exchange"])
print("\nNotional distribution summary:")
print(notional.describe())
```

```python
# Plot T1: Trade size histogram
fig = plot_trade_size_histogram(df)
fig.show()
save_figure(fig, "T1_trade_size_histogram", REPORTS_DIR)
```

```python
_, volley_stats = detect_volleys(df)
vsizes = volley_size_distribution(volley_stats)

print("Volley size distribution:")
print(vsizes)
```

```python
# Plot T3: Volley distribution
fig = plot_volley_distribution(volley_stats)
fig.show()
save_figure(fig, "T3_volley_distribution", REPORTS_DIR)
```

```python
imbalance = compute_aggressor_imbalance(df)
buy_ratio = compute_overall_buy_ratio(df)
tick_freq = compute_tick_frequency(df)

print("Overall buy ratio per exchange:")
print(buy_ratio)
```

```python
# Plot T2: Aggressor imbalance
fig = plot_aggressor_imbalance(imbalance)
fig.show()
save_figure(fig, "T2_aggressor_imbalance", REPORTS_DIR)
```

```python
# Plot T4: Tick frequency
fig = plot_tick_frequency(tick_freq)
fig.show()
save_figure(fig, "T4_tick_frequency", REPORTS_DIR)
```
